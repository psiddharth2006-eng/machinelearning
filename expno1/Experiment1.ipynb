SCENARIO 1

```
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("/data.csv", encoding="ISO-8859-1")

print(df.head())
print(df.tail())
print(df.info())
print(df.describe())

print(df.isnull().sum())

df["TotalSales"] = df["Quantity"] * df["UnitPrice"]
df["InvoiceDate"] = pd.to_datetime(df["InvoiceDate"])

product_sales = df.groupby("Description")["TotalSales"].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,5))
product_sales.plot(kind="bar")
plt.show()

daily_sales = df.groupby(df["InvoiceDate"].dt.date)["TotalSales"].sum()

plt.figure(figsize=(10,5))
plt.plot(daily_sales)
plt.show()
```

```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("/diabetes.csv")

print(df.head())
print(df.info())
print(df.describe())

print((df == 0).sum())

plt.figure()
plt.hist(df["Glucose"], bins=20)
plt.show()

plt.figure()
plt.hist(df["Age"], bins=15)
plt.show()

plt.figure()
sns.boxplot(data=df[["Glucose", "BloodPressure", "BMI"]])
plt.show()
```

SCENARIO 3

```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("/Housing.csv")

print(df.head())
print(df.info())
print(df.describe())

print(df.isnull().sum())

categorical_cols_to_map = [
    'mainroad', 'guestroom', 'basement', 'hotwaterheating',
    'airconditioning', 'prefarea'
]

for col in categorical_cols_to_map:
    df[col] = df[col].map({'yes': 1, 'no': 0})

df['furnishingstatus'] = df['furnishingstatus'].map({'furnished': 2, 'semi-furnished': 1, 'unfurnished': 0})

plt.figure()
plt.scatter(df["area"], df["price"])
plt.show()

plt.figure(figsize=(8,6))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm")
plt.show()
```

```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("/marketing_campaign.csv", sep="\t")

print(df.head())
print(df.info())

print(df.isnull().sum())

df["Age"] = 2024 - df["Year_Birth"]

plt.figure()
plt.hist(df["Age"], bins=15)
plt.show()

plt.figure()
sns.boxplot(y=df["Income"])
plt.show()

spending_cols = [
    "MntWines", "MntFruits", "MntMeatProducts",
    "MntFishProducts", "MntSweetProducts", "MntGoldProds"
]

plt.figure(figsize=(10,5))
df[spending_cols].sum().plot(kind="bar")
plt.show()
```
